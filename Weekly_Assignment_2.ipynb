1. A Series is a one-dimensional labeled array is called series.
 DataFrame is a two-dimensional table with rows and columns is called dataframe.Both support operations like .head(), but .merge() works only on a DataFrame.

2. inplace=True modifies the original DataFrame instead of returning a new one.Pandas discourages its use because it can make code harder to understand and doesn't always improve performance
It's often better to write df = df.drop(...).

3. .fillna() replaces missing values with a specified value, while .dropna() removes rows or columns containing missing values. Use .fillna() when data is important to keep, and .dropna() when missing values make the data unreliable.

4. errors='coerce' converts invalid values to NaN instead of raising an error. This is useful for messy real-world data because it allows processing to continue without stopping due to invalid entries.

5. .loc selects data using labels, while .iloc selects data using integer positions. If the DataFrame index is not 0, 1, 2..., then df.loc[0] and df.iloc[0] may return different rows.

6. .map() is used for element-wise mapping in a Series, .apply() applies a function to each element or row, and .replace() substitutes specific values. Use .map() for dictionaries, apply() for custom functions, and replace() for value replacement.

7. This warning occurs when modifying a copy of a DataFrame instead of the original. For example, df[df["Age"]>20]["Age"]=30 triggers it. Fix it by using df.loc[df["Age"]>20, "Age"] = 30.

8.pd.cut() creates bins using fixed value ranges, while pd.qcut() creates bins with an equal number of observations. For annual income labels (Low/Mid/High), pd.qcut() is usually better because it distributes customers evenly across groups.

```python
import pandas as pd

df = pd.read_csv("indian_startups_fundingDADS(1).csv")

print("Rows and Columns:", df.shape)

print("\nMissing Values:")
print(df.isna().sum())

print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
print(sorted(df.columns))
```

```python
# Drop missing values
fy = df['founded_year'].dropna()
stats = {
    'count': fy.count(),
    'mean': fy.mean(),
    'min': fy.min(),
    'max': fy.max(),
    'median': fy.median(),
    'std': fy.std()
}

print(stats)
```

```python
def clean_amount(amount_str):
    if pd.isna(amount_str) or not isinstance(amount_str, str):
        return amount_str
    amount_str = amount_str.lower()
    amount_str = amount_str.replace('inr', '').replace('₹', '').replace(' ', '')
    if 'crore' in amount_str:
        amount_str = amount_str.replace('crore', '')
        try:
            return float(amount_str) * 10**7
        except ValueError:
            return pd.NA
    elif 'cr' in amount_str:
        amount_str = amount_str.replace('cr', '')
        try:
            return float(amount_str) * 10**7
        except ValueError:
            return pd.NA
    elif 'lac' in amount_str:
        amount_str = amount_str.replace('lac', '')
        try:
            return float(amount_str) * 10**5
        except ValueError:
            return pd.NA
    elif 'lakh' in amount_str:
        amount_str = amount_str.replace('lakh', '')
        try:
            return float(amount_str) * 10**5
        except ValueError:
            return pd.NA

    elif 'undisclosed' in amount_str:
        return pd.NA
    else:
        try:
            return float(amount_str)
        except ValueError:
            return pd.NA

df['amount'] = df['amount'].apply(clean_amount)
df['amount'] = pd.to_numeric(df['amount'])

print("dtype:", df['amount'].dtype)

print("NaN values:", df['amount'].isna().sum())

print("Mean funding amount:", df['amount'].mean())

print("Minimum funding amount:", df['amount'].min())
print("Maximum funding amount:", df['amount'].max())
```

```python
# Convert funding_date to datetime
df['funding_date'] = pd.to_datetime(df['funding_date'], format='mixed', errors='coerce')
year_counts = df['funding_date'].dt.year.value_counts().sort_index()

print("Deals in each year:")
print(year_counts)
print("\nYear with highest funding deals:")
print(year_counts.idxmax())
```

```python
# Standardise funding_stage
stage_map = {
    'seed': 'Seed',
    'pre-seed': 'Pre-Seed',
    'series a': 'Series A',
    'series-a': 'Series A',
    'series_a': 'Series A',
    'series b': 'Series B',
    'series-b': 'Series B',
    'series_b': 'Series B',
    'series c': 'Series C',
    'series-c': 'Series C'
}

df['funding_stage'] = (
    df['funding_stage']
      .str.strip()
      .str.lower()
      .replace(stage_map)
)
print(df['funding_stage'].value_counts())
```

```python
result = df[
    (df['city'] == 'Bengaluru') &
    (df['amount'] > 100)
][['startup_name', 'amount']]

result = result.sort_values(by='amount', ascending=False)

print(result)
```

```python
count = df[df['founded_year'].between(2015, 2020)].shape[0]

print("Total startups founded between 2015 and 2020:", count)
```

```python
result = df[
    df['startup_name'].str.contains('pay', case=False, na=False)
]['startup_name'].unique()

print(result)
```

```python
# Create funding_efficiency column
df['funding_efficiency'] = df['amount'] / (2024 - df['founded_year'])
top10 = df[['startup_name', 'industry', 'amount', 'founded_year', 'funding_efficiency']] \
    .sort_values(by='funding_efficiency', ascending=False) \
    .head(10)

print(top10)
```

```python
# Create funding_tier column
df['funding_tier'] = df['amount'].apply(
    lambda x: 'Unicorn' if pd.notna(x) and x >= 1000 else
              'Mega' if pd.notna(x) and x >= 500 else
              'Large' if pd.notna(x) and x >= 100 else
              'Mid' if pd.notna(x) and x >= 10 else
              'Small / Unknown'
)
print(df['funding_tier'].value_counts())
```

```python
import pandas as pd
df['era'] = pd.cut(
    df['founded_year'],
    bins=[0, 2009, 2017, float('inf')],
    labels=['Early', 'Growth', 'Recent']
)

print(df['era'].value_counts())
```

```python
# Create investor_count column
df['investor_count'] = df['investors'].apply(
    lambda x: 0 if pd.isna(x) else len(str(x).split(', '))
)

print(df['investor_count'].value_counts().sort_index())
```

```python
# Average funding amount per industry
avg_funding = (
    df.groupby('industry')['amount']
      .mean()
      .sort_values(ascending=False)
      .head(5)
)

print(avg_funding)
```

```python
city_stats = df.groupby('city').agg(
    number_of_startups=('startup_name', 'count'),
    total_funding=('amount', 'sum'),
    average_funding=('amount', 'mean')
)

city_stats = city_stats.sort_values(
    by='number_of_startups',
    ascending=False
)

print(city_stats.head(10))
```

```python
result = df.groupby('industry').agg(
    count=('startup_name', 'count'),
    average_funding=('amount', 'mean')
)

result = result[result['count'] >= 15]

result = result.sort_values(by='average_funding', ascending=False)

print(result.head(1))
```

```python
# Find top 5 industries by startup count
top5_industries = df['industry'].value_counts().head(5).index

result = pd.crosstab(
    df[df['industry'].isin(top5_industries)]['industry'],
    df[df['industry'].isin(top5_industries)]['funding_tier']
)

print(result)
```

```python
# Find the index of the startup with the highest funding in each industry
idx = df.groupby('industry')['amount'].idxmax()

result = df.loc[idx, ['startup_name', 'industry', 'city', 'amount']]

print(result.sort_values('industry'))
```

```python
# Filter emerging players
emerging_players = df[
    (df['founded_year'] >= 2019) &
    (df['amount'] >= 100) &
    (df['industry'].isin(['Fintech', 'Edtech', 'HealthTech']))
]

emerging_players = emerging_players.sort_values(
    by='amount',
    ascending=False
)

print(emerging_players[
    ['startup_name', 'industry', 'city', 'founded_year', 'amount']
])
```

```python
# Group by city
city_funding = df.groupby('city').agg(
    count=('startup_name', 'count'),
    total_funding=('amount', 'sum')
)

city_funding = city_funding[city_funding['count'] >= 20]

city_funding = city_funding.sort_values(
    by='total_funding',
    ascending=False
)

print(city_funding.head(1))
```

```python
# Count startups for each (city, industry) combination
market_saturation = (
    df.groupby(['city', 'industry'])
      .size()
      .reset_index(name='startup_count')
)
market_saturation = market_saturation.sort_values(
    by='startup_count',
    ascending=False
)

print(market_saturation.head(10))
```
